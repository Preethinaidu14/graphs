## GRAPH USING POLAR COORDINATES

```python
t = np.arange(0,2*np.pi,0.01)
n = 100
r = 1+2*np.log(10*n*t)
plt.polar(t,r)
plt.show()
```

```python
t = np.arange(0,2*np.pi,0.01)
for n in range(0,5):
    r = np.cos(10*n*t)
    plt.polar(t,r)
plt.show()
```

## MONTE CARLO METHOD

```python
from matplotlib import pyplot as plt
import numpy as np
from numpy import random

N = 1000

circle_x = []
circle_y = []

square_x = []
square_y = []

i = 1
while i<=N:
    x = random.uniform(-1,1)
    y = random.uniform(-1,1)
    if (x**2 + y**2 <= 1):
        circle_x.append(x)
        circle_y.append(y)
    else:
        square_x.append(x)
        square_y.append(y)
    i+=1
pi = 4*len(circle_x)/N
print("theoretical pi value=",np.pi)
print("through monte carlo pi value=", pi)

plt.plot(circle_x,circle_y,'r.')
plt.plot(square_x,square_y,'b.')
plt.grid()
plt.show()
```

## RADIUS OF CURVATURE

```python
from matplotlib import pyplot as plt
import numpy as np
t = np.linspace(0,2*np.pi,200)
x = np.sin(t)
y = np.sin(t)*np.cos(t)
dx = np.cos(t)
dy = np.cos(2*t)
ddx = -np.sin(t)
ddy = -2*np.sin(2*t)
rho = ((dx**2+dy**2)**3/2)/(dx*ddy-dy*ddx)
#p = 1/rho
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x,(1/rho))
plt.show()
```
